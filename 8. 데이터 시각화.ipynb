## 🧩 matplotlib.pyplot의 주요 기능

plt는 그래프를 그리는 함수 모음입니다.
아래 표는 생명과학 데이터 분석에서 자주 사용하는 plt 함수와 그 역할을 정리한 것입니다.

---

|        함수        | 의미                | 예시 코드                     | 설명                    |
| :--------------: | :---------------- | :------------------------ | :-------------------- |
|   `plt.plot()`   | 선 그래프(Line Plot)  | `plt.plot(x, y)`          | 시간에 따른 변화나 경향을 표현     |
|    `plt.bar()`   | 막대그래프(Bar Plot)   | `plt.bar(x, y)`           | 처리군 간의 값 비교에 사용       |
|  `plt.scatter()` | 산점도(Scatter Plot) | `plt.scatter(x, y)`       | 두 변수 간 상관관계 확인        |
|   `plt.title()`  | 그래프 제목            | `plt.title("세포 성장 곡선")`   | 그래프 상단 제목 표시          |
|  `plt.xlabel()`  | X축 이름             | `plt.xlabel("시간 (hour)")` | X축의 의미 표시             |
|  `plt.ylabel()`  | Y축 이름             | `plt.ylabel("세포 수")`      | Y축의 의미 표시             |
| `plt.grid(True)` | 격자선 추가            | `plt.grid(True)`          | 값의 위치를 쉽게 파악          |
|  `plt.legend()`  | 범례 표시             | `plt.legend()`            | 여러 데이터의 구분 표시         |
|   `plt.show()`   | 그래프 출력            | `plt.show()`              | 그래프를 화면에 표시 (마지막에 필수) |

````python
# 데이터 시각화 라이브러리 불러오기
import matplotlib.pyplot as plt
````

````python
 # 실험 데이터
time = [0, 2, 4, 6, 8, 10]
cell_count = [10, 25, 45, 70, 100, 130]

# 선 그래프 그리기
plt.plot(time, cell_count, marker='o', color="teal")
plt.title("세포 성장 곡선")
plt.xlabel("시간 (hour)")
plt.ylabel("세포 수 (x10⁴)")
plt.grid(True)
plt.show()
````

````python
# 3가지 유전자의 발현량
genes = ["Gene_A", "Gene_B", "Gene_C"]
expression = [40, 70, 55]

# 막대 그래프 그리기
plt.bar(genes, expression, color=['skyblue', 'salmon', 'lightgreen'])
plt.title("유전자별 발현량 비교 (Gene Expression)")
plt.xlabel("유전자")
plt.ylabel("발현량 (TPM)")
plt.show()
````

```python

````python
# 세 가지 처리 조건의 유전자 발현 데이터
control = [45, 50, 52, 55, 58, 60, 62]       # 대조군
treatment_A = [60, 63, 65, 68, 70, 72, 75]   # 실험군 A
treatment_B = [35, 38, 40, 42, 45, 47, 50]   # 실험군 B

# 박스플롯 그리기
plt.boxplot([control, treatment_A, treatment_B],
            labels=['Control', 'Treatment A', 'Treatment B'])
plt.title("조건별 유전자 발현 분포")
plt.ylabel("발현량 (TPM)")
plt.grid(True)
plt.show()
````

```python

````python
# NumPy는 수학 계산을 위한 라이브러리
import numpy as np

# 재현성을 위해 시드 설정
np.random.seed(31)

# 100개의 가상 유전자 발현 데이터 생성
control = np.random.normal(loc=50, scale=10, size=100)            # 대조군 평균 50, 표준편차 10
treatment = control + np.random.normal(loc=3, scale=8, size=100)  # 실험군은 대조군 대비 평균 3 표준편차 8 증가

# 산점도(scatter) 그리기
plt.figure(figsize=(6, 5))
plt.scatter(control, treatment, color='skyblue', alpha=0.8, edgecolors='w', s=70)
plt.title("대조군 vs 처리군 유전자 발현 비교")
plt.xlabel("대조군 발현량 (TPM)")
plt.ylabel("처리군 발현량 (TPM)")
plt.grid(True)

# 대각선 추가 (변화 기준선)
plt.plot([20, 80], [20, 80], 'r--', label="기준선 (y=x)")
plt.legend()   # 범례 표시 -> 기준선 (y=x)
plt.show()
````

### ✅ 배운 내용
- `matplotlib`을 활용해 **실험 데이터의 시각화 그래프를 생성**할 수 있다.  
- **선 그래프(line plot)**, **막대 그래프(bar plot)**, **박스 플롯(boxplot)**, **산점도(scatter plot)** 를 그릴 수 있다.  
- 축 제목(`xlabel`, `ylabel`)과 그래프 제목(`title`)을 추가해 **데이터의 의미를 명확히 표현**할 수 있다.  
- `plt.legend()`와 `plt.grid()`를 이용해 **가독성을 높인 그래프**를 작성할 수 있다.  
- 생물 실험 데이터를 기반으로 **시간에 따른 변화나 비교 결과를 시각적으로 표현**할 수 있다.  

---

### 🧬 생명과학 응용
- **세포 성장 곡선 분석** — 시간에 따른 세포 수 증가를 시각화하여 성장률 비교  
- **온도·pH 영향 평가** — 온도나 pH 조건에 따른 효소 반응 속도 변화를 시각적으로 비교  
- **단백질 발현량 비교** — 실험군과 대조군 간의 발현 수준 차이를 막대그래프로 표현  
- **실험 결과 보고서 작성** — 데이터의 시각적 패턴을 활용해 연구 결론을 직관적으로 제시
