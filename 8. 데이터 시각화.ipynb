````python
# Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∂àÎü¨Ïò§Í∏∞
import matplotlib.pyplot as plt
````

````python
 # Ïã§Ìóò Îç∞Ïù¥ÌÑ∞
time = [0, 2, 4, 6, 8, 10]
cell_count = [10, 25, 45, 70, 100, 130]

# ÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.plot(time, cell_count, marker='o', color='teal')
plt.title("ÏÑ∏Ìè¨ ÏÑ±Ïû• Í≥°ÏÑ†")
plt.xlabel("ÏãúÍ∞Ñ (hour)")
plt.ylabel("ÏÑ∏Ìè¨ Ïàò (x10‚Å¥)")
plt.grid(True)
plt.show()
````

````python
genes = ["Gene_A", "Gene_B", "Gene_C"]
expression = [40, 70, 55]

plt.bar(genes, expression, color=['skyblue', 'salmon', 'lightgreen'])
plt.title("Ïú†Ï†ÑÏûêÎ≥Ñ Î∞úÌòÑÎüâ ÎπÑÍµê (Gene Expression)")
plt.xlabel("Ïú†Ï†ÑÏûê")
plt.ylabel("Î∞úÌòÑÎüâ (TPM)")
plt.show()
````

```python

````python
# ÏÑ∏ Í∞ÄÏßÄ Ï≤òÎ¶¨ Ï°∞Í±¥Ïùò Ïú†Ï†ÑÏûê Î∞úÌòÑ Îç∞Ïù¥ÌÑ∞
control = [45, 50, 52, 55, 58, 60, 62]
treatment_A = [60, 63, 65, 68, 70, 72, 75]
treatment_B = [35, 38, 40, 42, 45, 47, 50]

# Î∞ïÏä§ÌîåÎ°Ø Í∑∏Î¶¨Í∏∞
plt.boxplot([control, treatment_A, treatment_B],
            labels=['Control', 'Treatment A', 'Treatment B'])
plt.title("Ï°∞Í±¥Î≥Ñ Ïú†Ï†ÑÏûê Î∞úÌòÑ Î∂ÑÌè¨")
plt.ylabel("Î∞úÌòÑÎüâ (TPM)")
plt.grid(True)
plt.show()
````

````python
# NumPyÎäî ÏàòÌïô Í≥ÑÏÇ∞ÏùÑ ÏúÑÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨
import numpy as np

# Ïû¨ÌòÑÏÑ±ÏùÑ ÏúÑÌï¥ ÏãúÎìú ÏÑ§Ï†ï
np.random.seed(42)

# 100Í∞úÏùò Í∞ÄÏÉÅ Ïú†Ï†ÑÏûê Î∞úÌòÑ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
control = np.random.normal(loc=50, scale=10, size=100)            # ÎåÄÏ°∞Íµ∞ ÌèâÍ∑† 50, ÌëúÏ§ÄÌé∏Ï∞® 10
treatment = control + np.random.normal(loc=3, scale=8, size=100)  # Ï≤òÎ¶¨Íµ∞ÏùÄ ÏïΩÍ∞Ñ Ï¶ùÍ∞Ä (ÌèâÍ∑† +3)

# ÏÇ∞Ï†êÎèÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(6, 5))
plt.scatter(control, treatment, color='skyblue', alpha=0.7, edgecolors='w', s=70)
plt.title("ÎåÄÏ°∞Íµ∞ vs Ï≤òÎ¶¨Íµ∞ Ïú†Ï†ÑÏûê Î∞úÌòÑ ÎπÑÍµê (100Í∞ú Ïú†Ï†ÑÏûê)")
plt.xlabel("ÎåÄÏ°∞Íµ∞ Î∞úÌòÑÎüâ (TPM)")
plt.ylabel("Ï≤òÎ¶¨Íµ∞ Î∞úÌòÑÎüâ (TPM)")
plt.grid(True)

# ÎåÄÍ∞ÅÏÑ† Ï∂îÍ∞Ä (Î≥ÄÌôî Í∏∞Ï§ÄÏÑ†)
plt.plot([20, 80], [20, 80], 'r--', label="Í∏∞Ï§ÄÏÑ† (y=x)")
plt.legend()
plt.show()
````
