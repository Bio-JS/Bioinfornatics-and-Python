##🧠 while 반복문 – 조건에 따른 반복

파이썬에서 while 반복문은 조건이 참(True)인 동안 코드 블록을 여러 번 실행할 때 사용됩니다.
이는 데이터 측정이 특정 임계값에 도달할 때까지, 혹은 실험 장비가 안정화될 때까지
반복적으로 동작해야 하는 생명과학 데이터 처리 과정에서 매우 유용합니다.

##✅ while 반복문

while 반복문은 조건식(Condition) 을 평가하여,
그 결과가 True일 때 블록 안의 코드를 계속 실행합니다.
조건이 False가 되면 반복을 멈춥니다.

```python
while 조건식:
    실행할 코드

## `while` 반복문 - 시퀀스 반복

````python
# 10줄의 염기서열을 순차적으로 읽는다고 가정하고 반복문을 작성하면 아래와 유사하다
# 아래 코드를 실행하고 2초 이내에 정지 버튼을 누르고 결과를 확인하시오.

counter = 0
while counter <= 10:
  print(counter)
  counter += 1
````

````python
# while 반복문에서 조건을 만족하지 않을 경우에는 else를 추가하면 된다.

counter = 0
while counter == 10:
  print(counter)
  counter += 1
else:
  print("end")
````

````python
# 세균 증식 시뮬레이션
# while 문을 이용하여 세균 수가 10,000에 도달할 때까지 두 배로 증가시키는 코드

bacteria_count = 100    # 초기 세균 수
generation = 0          # 세대 수 카운트

print("\n세균 증식 시뮬레이션:")

while bacteria_count < 10000:
    generation += 1             # 세대 수 증가
    bacteria_count *= 2         # 세균 수 두 배로 증가
    print(f"{generation}세대: 세균 수 = {bacteria_count:,} 마리")

print(f"{generation}세대 만에 세균 수가 {bacteria_count:,} 마리에 도달했습니다.")
````

## break – 반복을 즉시 종료하기

`break` 문은 반복문을 강제로 멈추는 명령어입니다.

`while` 반복문 내부에서 실행되면,
조건이 여전히 `True`여도 즉시 반복이 종료됩니다.

---

```python
✅ `while - break`의 기본 문법
while 조건:
    실행할 코드
    if 종료조건:   # 특정 상황 발생 시
        break      # 반복 즉시 종료


````python
# DNA 반보존적 복제 시뮬레이션
# 세포가 분열할 때마다 DNA 사본의 수가 두 배로 늘어나는 과정을 while 반복문으로 표현합니다.
# 목표 복제 수에 도달하면 break로 반복을 즉시 종료합니다.

dna_copies = 1      # 초기 DNA 사본 수 (1쌍)
generation = 0      # 세대 수
target_copies = 64  # 목표 복제 수

print("세대별 DNA 복제 과정")

while dna_copies > 0:      # 조건: DNA 사본이 0보다 큰 동안 반복 (내부 break로 종료)
    print(f"{generation}세대 후: DNA 사본 = {dna_copies}")
    generation += 1
    dna_copies *= 2        # 매 세대마다 복제 → 2배 증가

    if dna_copies >= target_copies:       # break 조건: 목표 복제 수에 도달하면 반복 종료
        print(f"\n⚠️ 목표 복제 수({target_copies}쌍)에 도달했습니다. 반복을 종료합니다.")
        break

print(f"총 {generation}세대 후, DNA 사본 수는 {dna_copies}쌍이 되었습니다.")
````

# **while 반복문으로 GC contents 계산하기**

* [HBB - subunit beta](https://www.ncbi.nlm.nih.gov/nuccore/AF181832.1?report=fasta)

````python
# HBB 서열을 입력하고 전체 뉴클레오타이드 길이를 확인하기

seq = input("서열을 입력하세요: ")
seq_len = len(seq)
print(seq_len)
````

````python
# GC_content를 구하기 위한 변수 생성

GC_content = 0
index = 0

# GC_content 계산을 위한 while 반복문

while index < seq_len:
  seq_index = seq[index]
  if seq_index == "G" or seq_index == "C":
    GC_content += 1
  index += 1              # index 들여쓰기에 주의, if문 안에 index가 있을 경우
else:                     # "A" or "T"가 입력되면 False가 되어 index += 1이 실행되지 않음
  print(f"GC함량은: {(GC_content / seq_len) * 100:.2f} %")
````
