## 비교 연산자 – 데이터의 관계를 판단하자!

파이썬의 **비교(Comparison) 연산자**는 두 값을 **비교하여 True(참)** 또는 **False(거짓)** 을 반환합니다.  
이 연산자들은 실험 결과나 생명과학 데이터에서 **값의 동일성, 크기, 기준 충족 여부**를 판단할 때 자주 사용됩니다.  

---

### ✅ **비교 연산자의 종류**

| 연산자 | 의미 | 예시 | 결과 |
|:--:|:--|:--|:--|
| `==` | 같다 | `5 == 5` | `True` |
| `!=` | 같지 않다 | `5 != 3` | `True` |
| `>` | 크다 | `7 > 4` | `True` |
| `<` | 작다 | `2 < 9` | `True` |
| `>=` | 크거나 같다 | `5 >= 5` | `True` |
| `<=` | 작거나 같다 | `3 <= 4` | `True` |

---

## 🧠 **비교 연산자의 특징**

- 모든 비교식의 결과는 **True** 또는 **False** 중 하나입니다.  
- 비교 연산은 **숫자, 문자열, 논리값** 등 다양한 자료형에 적용할 수 있습니다.  
- `True`는 **1**, `False`는 **0**으로 처리되므로, **산술 계산에도 활용**할 수 있습니다.  
- 여러 조건을 **논리 연산자(`and`, `or`, `not`)** 와 함께 조합하면 복합 판정을 자동화할 수 있습니다.  

---

## 🧬 **비교 연산자의 생명과학적 활용**

- **염기 서열 비교**: 두 염기가 같은지(`==`), 다른지(`!=`) 확인  
- **DNA 농도 확인**: 기준값 이상(`>=`)인지 판정  
- **배양 조건 점검**: 온도나 pH가 허용 범위 내(`<`, `>`)인지 확인  
- **품질 평가**: 흡광도 비율이 기준 이상(`>= 1.8`)인지 판정  



````python
# 정상 서열 vs 변이 서열의 불일치율(%) 계산

normal_sequence  = "ATGCTACCGTAA"
variant_sequence = "ATGCTATCGTGATTT"

# 두 서열의 길이 중 더 짧은 쪽을 기준으로 비교
comparison_length = min(len(normal_sequence), len(variant_sequence))
mismatch_count = 0

# 각 위치의 염기가 다른지 비교 (True=1, False=0)
for i in range(comparison_length):
    mismatch_count += (normal_sequence[i] != variant_sequence[i])

# 불일치율(%) 계산 → (불일치 개수 / 비교 길이) × 100
mismatch_percentage = mismatch_count / comparison_length * 100

# 결과 출력
print("정상 서열:", normal_sequence)
print("변이 서열:", variant_sequence)
print("비교 길이:", comparison_length)
print("불일치 개수:", mismatch_count)
print(f"불일치 비율: {mismatch_percentage:.1f}%")
````

## **논리(Logical) 연산자 – 여러 조건을 한 번에 판단하자!**

파이썬의 **논리 연산자(Logical Operators)** 는 여러 개의 **비교식이나 조건식**을  
**하나의 문장으로 결합할 때** 사용됩니다.  
즉, “이 조건도 만족하고(and), 저 조건도 만족한다(or)”처럼 **복합 판단**을 표현할 수 있습니다.  

---

### ✅ **논리 연산자의 종류**

| 연산자 | 의미 | 예시 | 결과 |
|:--:|:--|:--|:--|
| **`and`** | 두 조건이 모두 True일 때 True | `(A > 5) and (B < 10)` | `True` |
| **`or`** | 두 조건 중 하나라도 True면 True | `(A > 5) or (B < 10)` | `True` |
| **`not`** | 결과를 반대로 바꿈 | `not (A > 5)` | `False` |

---

### 🧠 **논리 연산자의 특징**

- 여러 조건을 **동시에 검사**할 때 사용합니다.  
- **and** → 모든 조건이 참이어야 True  
- **or** → 하나라도 참이면 True  
- **not** → 조건의 True/False를 반전시킵니다.  
- **비교 연산자**와 함께 사용하면 복합 판정식을 간단히 표현할 수 있습니다.  

---

### 🧬 **생명과학적 활용 예시**

- 실험 조건이 **온도·pH·산소 농도** 기준을 모두 만족하는지(`and`) 확인  
- 시료가 **농도나 순도 중 하나라도** 기준 미달인지(`or`) 판정  
- **대조군이 아닌 경우(`not`)** 에만 분석 실행  


````python
# 🧪 흡광도와 농도 분석을 통한 DNA 품질 판정

OD = 1.93          # 순도 지표 (A260/A280)
dna_conc = 35.0     # DNA 농도 (ng/µL)

# 조건 정의
pure = 1.8 <= OD <= 2.0               # 순도 기준 통과 여부
concentrated = dna_conc >= 20         # 농도 기준 통과 여부

# 논리 연산자를 이용한 종합 판정
if pure and concentrated:
    print("✅ DNA 품질 양호 (순도·농도 OK)")
elif pure and not concentrated:
    print("⚠️ 순도 OK, 농도 부족 → 농축 필요")
elif not pure and concentrated:
    print("⚠️ 농도 OK, 순도 불량 → 재정제 필요")
else:
    print("❌ 순도·농도 모두 기준 미달")
````

## 💾 **할당(Assignment) 연산자 – 변수의 값을 저장하고 갱신하자!**

파이썬의 **할당 연산자(Assignment Operators)** 는 변수를 **저장하거나 누적·갱신할 때** 사용됩니다.  
즉, 한 번 계산한 값을 다시 사용할 수 있고, 반복 실험처럼 **데이터가 누적되는 상황**을 표현할 수 있습니다.  

---

## 🧬 GC Skew란 무엇인가?

GC skew는 **DNA 이중가닥에서 G(구아닌)과 C(사이토신)의 비대칭 정도**를 수치로 표현한 값입니다.  

# GC skew는 DNA의 복제원점(Replication origin)을 찾을 때 사용됨


---

### **📌 GC Skew 공식**
GC_skew = (G - C) / (G + C)

- \( G \): 특정 구간에서 나타나는 구아닌(G)의 개수  
- \( C \): 특정 구간에서 나타나는 사이토신(C)의 개수  

### **📌 해석**
- GC Skew > 0 → **G가 C보다 많다**  
- GC Skew < 0 → **C가 G보다 많다**  
- GC Skew = 0 → G와 C의 비율이 동일  


[Proksee 접속하기](https://proksee.ca/)

````python
# 🧬 GC skew 변화 예시 – 중간에 방향이 바뀌는 서열
# GC skew는 DNA의 복제원점(Replication origin)을 찾을 때 사용됨

sequence = "GGGGGGGGGCCCCCCCCCCCCCCCCCCCCCGGGGGGGGGGGG"

G_count = 0
C_count = 0
index = 0

while index < len(sequence):
    base = sequence[index]

    if base == "G":
        G_count += 1     # G가 나오면 +1
    elif base == "C":
        C_count += 1     # C가 나오면 +1

    # GC_skew = (G - C) / (G + C)
    if (G_count + C_count) != 0:
        GC_skew = (G_count - C_count) / (G_count + C_count)
        print(f"{index+1}번째 염기까지 GC_skew: {GC_skew:.2f}")

    index += 1
````
