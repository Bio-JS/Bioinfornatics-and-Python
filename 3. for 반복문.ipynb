## 🧠 파이썬에서의 `for` 반복문 – 행동의 반복

## ✅ `for` 반복문

`for` 반복문은 다음과 같은 객체를 순회(iterate)할 때 사용됩니다.

- 리스트 (List)

- 범위 (Range)
- 그 외 반복 가능한 객체 (Iterable Objects)

## `for` 반복문 - 시퀀스 반복


````python
# 여러 개의 세균 샘플 처리하기
sample_list = ["bac1", "bac2", "bac3", "bac4", "bac5"]

print("실험 진행 과정")
for sample in sample_list:
    print(f"분석중인 샘플: {sample}")
````

````python
# DNA 서열 분석
dna_sequence = "ATCGATCGTAG"

print(f"분석 중인 서열: {dna_sequence}")
# enumerate는 위치(index)와 요소(value)를 동시에 꺼내주는 함수
for i, nucleotide in enumerate(dna_sequence):
    print(f"위치 {i+1}: {nucleotide}")
````

````python
# 실험 데이터 처리
cell_counts = [1000, 2000, 3000, 4000, 5000]

print("Cell count analysis:")
total_cells = 0
for count in cell_counts:
    total_cells += count  # total_cells = 1000
    print(f"샘플: {count} cells")

average_count = total_cells / len(cell_counts)    # len(cell_counts)는 요소의 개수
print(f"Average cell count: {average_count:,.0f} cells")
````

# `range()` 함수


## 🔢 파이썬 `range()` 함수 – 숫자 시퀀스 생성하기

`range()` 함수는 **일정한 규칙을 가진 숫자들의 나열(시퀀스)** 을 만들어낼 때 사용됩니다.  
특히 `for` 반복문에서 자주 활용됩니다.

---

### ✅ `range()`의 기본 문법

```python
range(start, stop, step)


````python
# 긴 DNA 서열 중 앞의 10개 염기만 출력하기

dna_sequence = "ATGCGTACGTTAGCCTGATCGGATCAGTACGATGCTAGCTAGCTAGCATGCA"

print("DNA 서열의 앞 10개 염기:")
for i in range(10, 20):  # 0부터 9까지 반복 → 총 10개 염기
    print(f"염기 {i+1}번: {dna_sequence[i]}")
````

## ✅ 개념 정리

| 구문 | 포함 여부 | 실제 범위 | 결과 값 |
|------|------------|------------|-----------|
| `range(0, 5)` | 0 포함, 5 제외 | 0~4 | 0, 1, 2, 3, 4 |
| `range(1, 6)` | 1 포함, 6 제외 | 1~5 | 1, 2, 3, 4, 5 |
| `range(11, 20)` | 11 포함, 20 제외 | 11~19 | 11, 12, …, 19 |

````python
# 🧬 정상 헤모글로빈과 변이 헤모글로빈 서열 비교

from Bio import SeqIO

# 두 개의 FASTA 파일 불러오기
normal_hbb = SeqIO.read("/content/sequence.fasta", "fasta")
variant_hbb = SeqIO.read("/content/sequence (1).fasta", "fasta")

# 염기서열만 추출
normal_hbb_seq = normal_hbb.seq
varient_hbb_seq = variant_hbb.seq

# 서로 다른 부분 찾아 출력하기
for i in range(0, 288):
    if normal_hbb_seq[i] == varient_hbb_seq[i]:   # != 는 같지 않다는 연산자
        print(f"{i}번째 염기와 {normal_hbb_seq[i]} 와 {varient_hbb_seq[i]} 일치한다.")
````
